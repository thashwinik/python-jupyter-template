### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 11)
y = x**2

# Plot
plt.plot(x, y, marker='o')

# Customizations
plt.title('Square of x from 0 to 10')
plt.xlabel('x')
plt.ylabel('x squared')
plt.grid(True)

# Show plot
plt.show()
```
