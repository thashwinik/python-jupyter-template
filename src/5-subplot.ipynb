### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Data for plots
x = np.arange(0, 11)
y_line = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Line plot (top-left)
axs[0, 0].plot(x, y_line, marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('x squared')
axs[0, 0].grid(True)

# Bar chart (top-right)
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot (bottom-left)
axs[1, 0].scatter(x_scatter, y_scatter)
for xi, yi in zip(x_scatter, y_scatter):
    axs[1, 0].annotate(f'({xi},{yi})', (xi, yi), textcoords="offset points", xytext=(5,5), ha='center')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].grid(True)

# Pie chart (bottom-right)
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')

plt.tight_layout()
plt.show()
```
